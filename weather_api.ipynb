### Status Code
Refere-se ao código de status HTTP, que indica o resultado de uma requisição feita ao servidor (se foi bem-sucedida, redirecionada, se houve erro do cliente ou erro do servidor).

| Classe  | Descrição        | Explicação                                                     |
| ------- | ---------------- | -------------------------------------------------------------- |
| **1xx** | Informacional    | A requisição foi recebida e o processo continua.               |
| **2xx** | Sucesso          | A requisição foi recebida, entendida e processada com sucesso. |
| **3xx** | Redirecionamento | É necessário realizar outra ação para completar a requisição.  |
| **4xx** | Erro do Cliente  | A requisição contém erro ou não pode ser processada.           |
| **5xx** | Erro do Servidor | O servidor falhou ao processar uma requisição válida.          |


````python
api_key = "bc682459572f4366975140119261702"
link_api = "http://api.weatherapi.com/v1//current.json"

parametros = {
    "key": api_key,
    "q": "Barra Velha",
    "lang": "pt"
}

resposta = requests.get(link_api,
                       params = parametros)

if resposta.status_code == 200:
    dados_requisicao = resposta.json()
    pprint.pprint(dados_requisicao)
    temp = dados_requisicao["current"]["temp_c"]
    descricao = dados_requisicao["current"]["condition"]["text"]
````
