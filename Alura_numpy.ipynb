# NumPy: análise numérica eficiente com Python

Atividades referentes ao curso da Alura (https://cursos.alura.com.br/course/numpy-analise-numerica-eficiente-pythons).

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Importando os dados
url_apples = "https://raw.githubusercontent.com/alura-cursos/numpy/dados/apples_ts.csv"
url_citrus = "https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv"

dados_apples = np.loadtxt(
    url_apples,
    delimiter=",",
    usecols=np.arange(1, 88, 1)
)

dados_citrus = np.loadtxt(
    url_citrus,
    skiprows=1,
    delimiter=",",
    usecols=np.arange(1, 6, 1)
)
````

````python
# Dimensão dos dados

print(f"Dimensão de dados_apples: {dados_apples.ndim}")
print(f"Dimensão de dados_citrus: {dados_citrus.ndim}")
````

````python
# Quantidade total dos elementos

print(f'Quantidade total de elementos de dados_apples: {dados_apples.size}')
print(f'Quantidade total de elementos de dados_citrus: {dados_citrus.size}')
````

````python
# Formato dos dados

print(f'Formato de dados_apples: {dados_apples.shape[0]} linhas e {dados_apples.shape[1]} colunas.')
print(f'Formato de dados_citrus: {dados_citrus.shape[0]} linhas e {dados_citrus.shape[1]} colunas.')
````

````python
# Transpor a matriz
print(f'Transposta de dados_apples: \n{dados_apples.T}')
print(f'\nTransposta de dados_citrus: \n{dados_citrus.T}')
````

````python
# Comparações com equal, allclose, all.

A = np.array([[1, 2, 3], [4.00001, 5, 6]])
B = np.array([[1, 2.00001, 3], [4, 5, 6]])

equal = np.equal(A,B)
print(f'Equal \n{equal}')
print('\n')

allclose = np.allclose(A,B)
print(f'Allclose \n{allclose}')
print('\n')

all = np.all(equal)
print(f'All \n{all}')
````

Crie arrays específicos para guardar o diâmetro e peso da laranja e toranja. O diâmetro está na coluna zero e o peso na coluna 1. Os dados referentes a laranja vão até a linha 4999 e os referentes à toranja iniciam na linha 5000 do arquivo.

## Atividade 2

Thaís queria desenhar um círculo com a ajuda do computador. Para desenhar a parte superior do círculo ela sabia que deveria usar a equação:

$
y = sqrt(1 - x ^ 2)
$

E para a parte inferior bastava utilizar a mesma equação multiplicada por -1.

Qual sequência de código ela deve usar para construir o gráfico corretamente usando a Numpy e sabendo que ambas as partes do círculo devem ter a mesma cor?

````python
# Gerar uma sequência de valores de x de -1 a 1
x = np.arange(-1, 1.01, 0.0001)

# Implementação da fórmula
y1 = np.sqrt(1 - x**2)
y2 = -np.sqrt(1 - x**2)

# Plotar o gráfico com as duas partes do círculo
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r' )

# Adicionar o título do gráfico e os rótulos dos eixos x e y
plt.title("Círculo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
````
