**Análise Geral de Compras**

* Número de itens exclusivos
* Preço médio de compra

```python
# Número de itens exclusivos
data.nunique()

```

```python
# Preço médio de compra
preco_medio = data['Valor'].mean().round(2)

print(f'O preço médio do valor é de R${preco_medio} reais.')
```

```python
# Número total de compras
total_compras = data['Valor'].count()

print(f'O número total de compras é de {total_compras}.')
```

```python
# Rendimento total

renda = data['Valor'].sum()

print(f'O rendimento total é de R${renda} reais.')
```

**Informações Demográficas Por Gênero**

* Porcentagem e contagem de compradores masculinos
* Porcentagem e contagem de compradores do sexo feminino
* Porcentagem e contagem de outros / não divulgados



```python
# Porcentagem e contagem de compradores masculinos

contagem_masc = data['Sexo'].value_counts()['Masculino']
porcentagem_masc = (contagem_masc/total_compras * 100).round(2)

print(f"A contagem masculina corresponde a {contagem_masc}, isso representa {porcentagem_masc}% da amostragem.")
```

```python
# Porcentagem e contagem de compradores do sexo feminino

contagem_fem = data['Sexo'].value_counts()['Feminino']
porcentagem_fem = (contagem_fem/total_compras *100).round(2)

print(f'A contagem feminina corresponde a {contagem_fem}, isso representa {porcentagem_fem}% da amostragem.')
```

```python
# Porcentagem e contagem de outros / não divulgados

data['Sexo'].unique() # Descobrindo como está escrito corretamente o nome na coluna

contagem_outro = data['Sexo'].value_counts()['Outro / Não Divulgado']
porcentagem_outro = (contagem_outro/total_compras *100).round(2)

print(f'A contagem de "Outro / Não Divulgado" corresponde a {contagem_outro}, isso representa {porcentagem_outro}% da amostragem.')
```

**Análise de Compras Por Gênero**

* Número de compras
* Preço médio de compra
* Valor Total de Compra
* Compras for faixa etária



```python
# Número de compras
compras_masc = data[data['Sexo'] == 'Masculino']['Valor'].count()
compras_fem = data[data['Sexo'] == 'Feminino']['Valor'].count()
compras_outro = data[data['Sexo'] == 'Outro / Não Divulgado']['Valor'].count()

print(f'O número de compras do gênero masculino é de {compras_masc}.')
print(f'O número de compras do gênero feminino é de {compras_fem}.')
print(f'O número de compras do gênero "Outro / Não Divulgado" é de {compras_outro}.')
```

```python
# Preço médio de compra
preco_medio_masc = data[data['Sexo'] == 'Masculino']['Valor'].mean().round(2)
preco_medio_fem = data[data['Sexo'] == 'Feminino']['Valor'].mean().round(2)
preco_medio_outro = data[data['Sexo'] == 'Outro / Não Divulgado']['Valor'].mean().round(2)

print(f'O número de preço médio do gênero masculino foi de R${preco_medio_masc} reais.')
print(f'O número de preço médio do gênero feminino foi de R${preco_medio_fem} reais.')
print(f'O número de preço mfoidio do gênero "Outro / Não Divulgado" foi de R${preco_medio_outro} reais.')
```

```python
# Valor Total de Compra
total_compras_masc = data[data['Sexo'] == 'Masculino']['Valor'].sum().round(2)
total_compras_fem = data[data['Sexo'] == 'Feminino']['Valor'].sum().round(2)
total_compras_outro = data[data['Sexo'] == 'Outro / Não Divulgado']['Valor'].sum().round(2)

print(f'O valor total de compras do gênero masculino foi de R${total_compras_masc} reais.')
print(f'O valor total de compras do gênero feminino foi de R${total_compras_fem} reais.')
print(f'O valor total de compras do gênero "Outro / Não Divulgado" foi de R${total_compras_outro} reais.')
```

```python
# Compras por faixa etária
compras_crianca = data[data['Idade'] < 10]['Valor'].count()
compras_adolescente = data[(data['Idade'] >= 10) & (data['Idade'] <= 17)]['Valor'].count()
compras_adulto = data[(data['Idade'] >= 18) & (data['Idade'] <= 59)]['Valor'].count()
compras_idosos = data[data['Idade'] >= 60]['Valor'].count()

print(f'O número de crianças que compraram corresponde a {compras_crianca}.')
print(f'O número de adolescentes que compraram corresponde a {compras_adolescente}.')
print(f'O número de adultos que compraram corresponde a {compras_adulto}.')
print(f'O número de idosos que compraram corresponde a {compras_idosos}.')
```

**Identifique os 5 principais compradores pelo valor total de compra e, em seguida, liste (em uma tabela):**

* Login
* Número de compras
* Preço médio de compra
* Valor Total de Compra



```python
# Agrupa as compras por Login
top_compradores = (
    data.groupby('Login')

        # Essa função aplica cálculos sobre cada grupo de compradores
        .agg(
            **{ # ** permite nomear as colunas de saída diretamente no agg()
                'Número de compras': ('Valor', 'count'),     # Traz a quantidade de compras
                'Preço médio de compra': ('Valor', 'mean'),  # Traz a média de compra
                'Valor Total de Compra': ('Valor', 'sum')    # Valor total da compra
              }
            )
        )

# Seleciona os 5 maiores compradores pelo valor total
top_compradores = (
    top_compradores

        # ordena do maior valor total para o menor
        .sort_values('Valor Total de Compra', ascending=False)
        # pega os 5 maiores
        .head(5)
)

# Formata os valores monetários
top_compradores['Preço médio de compra'] = top_compradores['Preço médio de compra'].map('${:.2f}'.format)
top_compradores['Valor Total de Compra'] = top_compradores['Valor Total de Compra'].map('${:.2f}'.format)

top_compradores
```

**Identifique os 5 itens mais populares por contagem de compras e, em seguida, liste (em uma tabela):**

* ID do item
* Nome do item
* Número de compras
* Preço do item
* Valor Total de Compra
* Itens mais lucrativos


```python
# Agrupa os dados por ID e Nome do item
itens_populares = (
    data.groupby(['Item ID', 'Nome do Item'])

    # Essa função aplica cálculos sobre cada grupo de compradores
    .agg(
        **{  # ** permite nomear as colunas de saída diretamente no agg()
            'Número de compras': ('Valor', 'count'),      # Conta quantas compras o item teve
            'Preço do item': ('Valor', 'mean'),           # Calcula o preço médio do item
            'Valor Total de Compra': ('Valor', 'sum')     # Soma o valor total vendido
        }
    )

    # Ordena do item mais comprado para o menos comprado
    .sort_values('Número de compras', ascending=False)

    # Seleciona apenas os 5 primeiros itens
    .head(5)
)

# Formata o preço do item com cifrão e duas casas decimais
itens_populares['Preço do item'] = itens_populares['Preço do item'].map('${:.2f}'.format)

# Formata o valor total de compra com cifrão e duas casas decimais
itens_populares['Valor Total de Compra'] = itens_populares['Valor Total de Compra'].map('${:.2f}'.format)

# Exibe a tabela final
itens_populares
```
